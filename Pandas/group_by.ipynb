# Data Import

```python
import pandas as pd
```

```python
order_leads = pd.read_csv('https://raw.githubusercontent.com/FBosler/Medium-Data-Exploration/master/order_leads.csv',
    parse_dates = [3])

sales_team = pd.read_csv(
    'https://raw.githubusercontent.com/FBosler/Medium-Data-Exploration/master/sales_team.csv',
    parse_dates = [3])
```

```python
# merges orders and sales
order_leads.head()
```

```python
sales_team.head()
```

```python
# merge data frames
df = pd.merge(order_leads, sales_team, on=['Company Id','Company Name'])
df = df.rename(columns = {'Order Value': 'Val', 'Converted': 'Sale'})
df.head()
```

# Grouping by Splitting `str.split()`

```python
df['Sales Rep'].str.split().str[0].unique()
```

```python
df.groupby(df['Sales Rep'].str.split(' ').str[0]).size()
```

# Grouping by whether or not there is a string is in the name of the rep

```python
df['Sales Rep'].apply(lambda x: 'William' in x) # returns True / False
# group by each 
df.groupby(df['Sales Rep'].apply(lambda x: 'William' in x)).size()
```

# Group by custom buckets  `pd.qcut()`, `pd.cut()`

```python
df.groupby(pd.qcut(df['Val'], q = 3, labels=['low','mid','high'])).size()
```

```python
df.groupby(pd.cut(df['Val'],[0,3000,5000,7000,10000])).size()
```

# `pd.Grouper()`

We can use different standard frequencies like 'D','W','M', or 'Q'. 'SM' for semi-month end frequency (15th and end of the month).

```python
df.groupby(pd.Grouper(key='Date',freq='Y')).size()
```

# Grouping by multiple columns

```python
df.groupby(['Company Name', 'Sales Rep']).size()
```

# `apply()`, `agg()`, `transform()` , `filter`

```python
df.groupby(pd.Grouper(key='Date',freq='Y'))['Val','Sale'].apply(sum)
```

```python
df.groupby('Sales Rep').agg({'Order Id':'size', 'Val':['sum','mean'], 'Sale':['sum','mean']})
```

# Named Aggregation

```python
import numpy as np

def cr(x):
    return round(np.mean(x),2)

aggregation = {'Potential Sales': ('Val','size'),
                'Sales': ('Sale','sum'),
                'Conversion Rate': ('Sale', cr)
}

df.groupby('Sales Rep').agg(**aggregation)
```

```python
df.groupby('Sales Rep')['Val'].transform(lambda x: x/sum(x)).round(3).head()
```

```python
# assigning results to a new column
df['Percent'] = df.groupby('Sales Rep')['Val'].transform(lambda x: x/sum(x)).round(3)
df.head()
```

# `filter()`

```python
df.groupby('Sales Rep').filter(lambda x: (x['Val'] * x['Sale']).sum() > 200000)
```
